```python
import random
import numpy as np

def k_means(dataset, max_iterations=100):
  # Initialize centroids randomly
  centroids = []
  for _ in range(5):
    centroids.append([random.randint(-75, -5), random.randint(-110, -50), random.randint(-95, -20)])

  # Initialize previous cluster assignments
  previous_assignments = [None] * len(dataset)

  for iteration in range(max_iterations):
    # Assign points to centroids
    clusters = [[] for _ in range(5)]
    current_assignments = [None] * len(dataset)

    for i, point in enumerate(dataset.values):
      min_distance = float('inf')
      closest_centroid_index = -1

      for j, centroid in enumerate(centroids):
        distance = np.sqrt((point[0] - centroid[0])**2 + (point[1] - centroid[1])**2 + (point[2] - centroid[2])**2)
        if distance < min_distance:
          min_distance = distance
          closest_centroid_index = j

      clusters[closest_centroid_index].append(point)
      current_assignments[i] = closest_centroid_index

    # Check for convergence (cluster assignments)
    if current_assignments == previous_assignments:
      break

    # Update centroids
    for i, cluster in enumerate(clusters):
      if cluster:
        centroids[i] = np.mean(cluster, axis=0).tolist()

    previous_assignments = current_assignments

  return clusters, centroids
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

clusters, centroids = k_means(dataframe)

# Assuming 'clusters' and 'centroids' are returned by k_means
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot clusters
for cluster in clusters:
    x = [point[0] for point in cluster]
    y = [point[1] for point in cluster]
    z = [point[2] for point in cluster]
    ax.scatter(x, y, z, marker='o')  # Use different markers for clusters

# Plot centroids
centroids_x = [centroid[0] for centroid in centroids]
centroids_y = [centroid[1] for centroid in centroids]
centroids_z = [centroid[2] for centroid in centroids]
ax.scatter(centroids_x, centroids_y, centroids_z, marker='x', s=100, c='red')  # Highlight centroids

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('K-means Clustering Results')

plt.show()
```
