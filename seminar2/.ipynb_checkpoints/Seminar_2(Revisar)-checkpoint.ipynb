### Skelleton points visualization function

```python
def SkelletonPoints(skelleton_coordinates, pose_labels, train_labels):
    '''
    Plots a visualization of a specific skeleton pose using keypoint coordinates and their connections.

    Parameters:
    skelleton_coordinates (pandas.DataFrame): A single-row DataFrame with 36 values:
                                               the first 18 are x-coordinates, the next 18 are y-coordinates.
    pose_labels (list or np.ndarray): List of human-readable pose labels (e.g., ['wave', 'push', ...]).
    train_labels (pandas.DataFrame): DataFrame with integer pose labels (1 to 5) matching pose_labels.

    Returns:
    tuple: (x, y) coordinates of the 18 skeleton keypoints as lists.
    '''

    connections = [
        [0, 1], [1, 2], [2, 3], [3, 4],
        [1, 5], [5, 6], [6, 7],
        [1, 8], [8, 9], [9, 10],
        [1, 11], [11, 12], [12, 13],
        [0, 14], [0, 15], [14, 16], [15, 17],
        [2, 5], [8, 11]
    ]

    x = []
    y = []
    for i in range(skelleton_coordinates.shape[1] - 18):
        x.append(skelleton_coordinates.iloc[0, i])
        y.append(skelleton_coordinates.iloc[0, i + 18])

    coordinates = [[x[i], y[i]] for i in range(18)]

    for i, (x_p, y_p) in enumerate(coordinates):
        plt.plot(y_p, x_p, 'o', color='blue')

    plt.gca().invert_yaxis()
    plt.axis('equal')

    for a, b in connections:
        x1, y1 = coordinates[a]
        x2, y2 = coordinates[b]
        dx, dy = x2 - x1, y2 - y1
        plt.arrow(y1, x1, dy, dx, head_width=0.1, length_includes_head=True, color='red')

   
    pos = skelleton_coordinates.index.to_list()[0]
    pose_index = int(train_labels.iloc[pos, 0])
    plt.title('Actual pose: ' + pose_labels[pose_index - 1])


    return x, y
```

```python
poses = np.array(['wave', 'push', 'crouch', 'sitdown', 'bend'])

values_skelleton = train_skelleton_points.iloc[:,:36]
random_skelleton_coordinates = values_skelleton.sample(n=1)
```

```python
x, y = SkelletonPoints(random_skelleton_coordinates, poses, train_labels)
```

# Testing

```python
# Crear la gráfica de dispersión
plt.figure(figsize=(6, 6))
plt.scatter(y, x, color='b', marker='o')

plt.gca().invert_yaxis()

# Etiquetas y título
plt.title("Gráfica de dispersión de puntos (x, y)")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.axis('equal')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

# **Part 1 - CSI plots**

## Importing data

```python
# Paths to CSI data:
file_path="/content/drive/MyDrive/4º/3r Trimestre/ML/S2/"
train_features = "data/Train_features.csv"
train_labels = "data/Train_labels.csv"
test_labels = "data/Test_labels.csv"
test_features = "data/Test_features.csv"

# CSI
# Load train_features into a pandas dataframe
dataframe = pd.read_csv(file_path+train_features, sep=r',', header=None)

dataframe.head()
```

## CSI averages & plot

```python
# CSI average

averages_samples = []
for i in range(1000):
  average_per_subcarrier = []
  for j in range(30):
    it=9*j
    average_per_subcarrier.append(np.mean(dataframe.iloc[i][it:it+8]))
  averages_samples.append(average_per_subcarrier)

averages_samples_df = pd.DataFrame(averages_samples)
averages_samples_df.head()
```

```python
# Plot of the first sample

plt.plot(averages_samples_df.iloc[0])
plt.title("Subcarriers sample 0")
plt.xlabel("Subcarrier")
plt.ylabel("CSI average")
plt.show()
```

## CSI averages per class & plot

```python
samples_class = pd.read_csv(file_path+train_labels, sep=r',', header=None)
averages_per_class = []
for i in range(5):
  averages_per_class.append([])
  for j in range(30):
    averages_per_class[i].append([])

for i in range(5):
  for j in range(1000):
    if samples_class.iloc[0][j] != i+1: continue
    for k in range(30):
      averages_per_class[i][k].append(averages_samples_df.iloc[j][k])

for i in range(5):
  for j in range(30):
    averages_per_class[i][j] = np.mean(averages_per_class[i][j])

averages_per_class_df = pd.DataFrame(averages_per_class)
averages_per_class_df.head()
```

```python
# Plot of the averages per subcarrier per class

for i in range(5):
  plt.plot(averages_per_class_df.iloc[i])
plt.title("Subcarriers per class")
plt.xlabel("Subcarrier")
plt.ylabel("CSI average")
plt.legend(["Wave", "Push", "Crouch", "Sitdown", "Bend"])
plt.show()
```

# Part 2 - SVM

```python
train_features_df = pd.read_csv(file_path+train_features, sep=r',', header=None)
train_labels_df = pd.read_csv(file_path+train_labels, sep=r',', header=None)
test_features_df = pd.read_csv(file_path+test_features, sep=r',', header=None)
test_labels_df = pd.read_csv(file_path+test_labels, sep=r',', header=None)

# Create and train the SVM model
svm_model = SVC(kernel='linear')  # You can experiment with different kernels
svm_model.fit(train_features_df, train_labels_df.values.ravel())

# Generate predictions on the test data
predictions = svm_model.predict(test_features_df)

# Evaluate the model using accuracy
accuracy = accuracy_score(test_labels_df.values.ravel(), predictions)
test_mse = mean_squared_error(predictions, test_labels_df.values.ravel())
test_r2 = r2_score(predictions, test_labels_df.values.ravel())

print(f"Test MSE: {test_mse:.2f}")
print(f"Test R² Score: {test_r2:.4f}")
print(f"SVM Accuracy: {accuracy}")
```
