## Part 1 - Data Preparation & Analysis

### Data preparation

```python
# Seminar 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Queremos predecir el Throughput

dataset = pd.read_csv('data/dataset_Seminar1.txt',sep=' ',header=None)
Features = ['n_STAs','load','sx','sy','area','CW',
            'channelw','packet_size','max_RSSI', 'avg_RSSI','min_RSSI','avg_prob_failure',
           'tput','avg_delay','tot_airtime','prop_airtime']
for i in range(len(Features)):
    dataset = dataset.rename(columns={i:Features[i]})
print(dataset.head())
```

```python
dataset.describe()
```

```python
# Gráficos
plt.figure(figsize=(12, 6))
sns.histplot(dataset['tput'], bins=30, kde=True)
plt.title('Distribución del Throughput')
plt.xlabel('Throughput (bps)')
plt.ylabel('Frecuencia')
plt.show()

# Gráficos
plt.figure(figsize=(12, 6))
sns.histplot(dataset['load'], bins=30, kde=True)
plt.title('Distribución de la Carga')
plt.xlabel('Load (Mbps)')
plt.ylabel('Frecuencia')
plt.show()

# Gráficos
plt.figure(figsize=(12, 6))
sns.histplot(dataset['tot_airtime'], bins=30, kde=True)
plt.title('Distribución del tiempo necesario')
plt.xlabel('Time (s)')
plt.ylabel('Frecuencia')
plt.show()
```
